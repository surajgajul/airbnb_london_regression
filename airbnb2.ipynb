```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
airbnb_og=pd.read_csv('london_weekends.csv')
airbnb_og.head()
```

```python
airbnb_og.drop("Unnamed: 0", axis=1, inplace=True)
airbnb_og
```

```python
airbnb_og.info()
```

```python
bool_cols=['room_shared', 'room_private', 'host_is_superhost']
airbnb_og[bool_cols]=airbnb_og[bool_cols].astype(int)
airbnb_og
```

```python
cat_col='room_type'
airbnb_encoded = pd.get_dummies(airbnb_og[cat_col], prefix=cat_col)
airbnb = pd.concat([airbnb_og, airbnb_encoded], axis=1)
airbnb = airbnb.drop(columns=cat_col)
airbnb
```

```python
corr_airbnb=airbnb.corr()
corr_airbnb['realSum'].sort_values(ascending=False)
```

```python
airbnb.drop(['cleanliness_rating', 'guest_satisfaction_overall', 'host_is_superhost', 'room_shared', 'multi', 'room_shared', 'room_private'], axis=1, inplace=True)
```

```python
airbnb
```

```python
airbnb_train=airbnb.copy()
airbnb_train.drop('realSum', inplace=True, axis=1)
airbnb_train
```

```python
airbnb_labels=airbnb['realSum'].copy()
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(airbnb_train, airbnb_labels)
```

```python
some_data = airbnb_train.iloc[:5]
some_labels = airbnb_labels.iloc[:5]
print("Predictions:", tree_reg.predict(some_data))
```

```python
print("Labels", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
airbnb_predictions = tree_reg.predict(airbnb_train)
tree_mse = mean_squared_error(airbnb_labels, airbnb_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, airbnb_train, airbnb_labels,
scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
display_scores(tree_rmse_scores)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
import xgboost as xgb
```

```python
param_grid = [
{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},]
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse')
random_search = RandomizedSearchCV(xgb_model, param_grid, n_iter=10,
scoring='neg_mean_squared_error',return_train_score=True)
random_search.fit(airbnb_train, airbnb_labels)
```

```python
random_search.best_params_
```

```python
feature_importances = random_search.best_estimator_.feature_importances_
feature_importances
```

```python
cvres = random_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
some_data = airbnb_train.iloc[:5]
some_labels = airbnb_labels.iloc[:5]
print("Predictions:", random_search.predict(some_data))
```

```python
print("Labels", list(some_labels))
```

```python
from catboost import CatBoostRegressor
catb_model=CatBoostRegressor(iterations=100, learning_rate=0.1, random_seed=42)
catb_model.fit(airbnb_train, airbnb_labels, verbose=False)
```

```python
some_data = airbnb_train.iloc[3:8]
some_labels = airbnb_labels.iloc[3:8]
print("Predictions:", catb_model.predict(some_data))
```

```python
print("Labels", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
predictions = catb_model.predict(some_data)
mse = mean_squared_error(some_labels, predictions)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
airbnb_catb=airbnb.copy()
airbnb_catb
```

```python
airbnb_catb_labels=airbnb_catb['realSum'].copy()
airbnb_catb.drop('realSum', axis=1, inplace=True)
airbnb_catb
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(airbnb_catb, airbnb_catb_labels, test_size=0.1, random_state=42)
```

```python
model = CatBoostRegressor(iterations=100, learning_rate=0.1, random_seed=42)
```

```python
model.fit(X_train, y_train, verbose=False)
```

```python
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

# Define the hyperparameter search space
param_dist = {'learning_rate': uniform(0.01, 0.1),
              'depth': [3, 4, 5, 6, 7, 8],
              'l2_leaf_reg': uniform(1, 10)}
```

```python
cb = CatBoostRegressor(verbose=False)

# Perform the randomized search
rand_search = RandomizedSearchCV(cb, param_distributions=param_dist, 
                                 n_iter=10, cv=5, n_jobs=-1, random_state=42)

# Fit the randomized search to the data
rand_search.fit(X_train, y_train)

# Print the best hyperparameters and score
print("Best Hyperparameters:", rand_search.best_params_)
print("Best Score:", rand_search.best_score_)
```

```python
import pandas as pd

# Extract the RMSE scores for each combination of hyperparameters
results = pd.DataFrame(rand_search.cv_results_)
scores = results[['params', 'mean_test_score', 'std_test_score']]

# Print the RMSE scores for each combination of hyperparameters
for i, score in scores.iterrows():
    print("Hyperparameters:", score['params'])
    print("Mean RMSE Score:", np.sqrt(score['mean_test_score']))
    print("Standard Deviation:", score['std_test_score'])
```

```python
some_data = airbnb_catb.iloc[3:8]
some_labels = airbnb_catb_labels.iloc[3:8]
print("Predictions:", rand_search.predict(some_data))
```

```python
print("Labels", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
predictions = rand_search.predict(some_data)
mse = mean_squared_error(some_labels, predictions)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```
