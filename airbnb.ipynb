```python
from sklearn.model_selection import train_test_split
airbnb_train_set, airbnb_test_set = train_test_split(airbnb, test_size=0.2,
random_state=42)
```

```python
airbnb=airbnb_train_set.copy()
airbnb_labels=airbnb['realSum'].copy()
```

```python
airbnb.drop('realSum', axis=1, inplace=True)
airbnb.head()
```

```python
airbnb_num=airbnb.copy()
airbnb_num.drop(['room_type', 'room_shared', 'room_private', 'host_is_superhost', 'multi', 'biz'], axis=1, inplace=True)
airbnb_num.head()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
X=imputer.fit_transform(airbnb_num)
```

```python
airbnb_tr=pd.DataFrame(X, columns=airbnb_num.columns, index=airbnb_num.index)
airbnb_tr.info()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
num_pipeline = Pipeline([
('imputer', SimpleImputer(strategy="median")),
('std_scaler', StandardScaler()),
])
airbnb_num_tr = num_pipeline.fit_transform(airbnb_num)
```

```python
airbnb.drop(['room_shared', 'room_private'], axis=1, inplace=True)
airbnb
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
num_attribs = list(airbnb_num)
cat_attribs = ["room_type"]
full_pipeline = ColumnTransformer([
("num", num_pipeline, num_attribs),
("cat", OneHotEncoder(), cat_attribs),
])
airbnb_prepared = full_pipeline.fit_transform(airbnb)
airbnb_prepared[0]
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor()
tree_reg.fit(airbnb_prepared, airbnb_labels)
```

```python
some_data = airbnb.iloc[:5]
some_labels = airbnb_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions:", tree_reg.predict(some_data_prepared))
```

```python
print("Labels", list(some_labels))
```

```python
from sklearn.metrics import mean_squared_error
airbnb_predictions = tree_reg.predict(airbnb_prepared)
tree_mse = mean_squared_error(airbnb_labels, airbnb_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, airbnb_prepared, airbnb_labels,
scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
display_scores(tree_rmse_scores)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
param_grid = [
{'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
{'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},]
forest_reg = RandomForestRegressor()
random_search = RandomizedSearchCV(forest_reg, param_grid, n_iter=10,
scoring='neg_mean_squared_error',return_train_score=True)
random_search.fit(airbnb_prepared, airbnb_labels)
```

```python
random_search.best_params_
```

```python
feature_importances = random_search.best_estimator_.feature_importances_
feature_importances
```

```python
cvres = random_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```
